```python
import urllib.request as request
import json

json_str = request.urlopen("https://api.github.com/repositories").read() # read() -> bytes 
print(type(json_str)) # bytes 객체
output = json.loads(json_str)
print(type(output)) # 딕셔너리가 모여있는 리스트 -> JSON

for item in output:
    print(item["name"])
    print(item["full_name"])
    print(item["owner"]["login"])
    print()
```

```python
import json
price = {
    "date" : "2019-04-02",
    "price" : {
        "apple" : 80,
        "orange" : 20,
        "banana" : 40
    }
}

# json.dumps()는 파이썬 객체(예제에서는 dictionary)를 json 형식의 문자열로 변환 -> string
# json.loads() -> dict
# json은 파이썬에서 dict형태로 표현됨

json_str = json.dumps(price)
print(type(json_str))
print(json_str)

data = json.loads(json_str)
print(type(data))
print(data["date"])
print(data["price"]["apple"])
```

```python
import csv, codecs

with codecs.open("test.csv", "w", "utf-8") as csvfile:
    writer = csv.writer(csvfile, delimiter = ",", quotechar = '"')
    
    # write method를 사용하여 한 줄씩 출력
    writer.writerow(["ID", "이름", "가격"])
    writer.writerow(["1000", "SD카드", 30000])
    writer.writerow(["1001", "키보드", "21000"])
    writer.writerow(["1002", "마우스", "15000"])

# utf-8로 저장할 경우 excel에서 읽지 못할 수 있음
```
