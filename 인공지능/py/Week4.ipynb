```python
import urllib.request
import urllib.parse
API = 'https://search.naver.com/search.naver'
values = {
    'sm' : 'top_hty',
    'fbm' : '1',
    'ie' : 'utf8',
    'query' : '초콜릿'
}
# 한글, 특수문자등을 자동으로 encoding
params = urllib.parse.urlencode(values)

url = API + '?' + params
print('url =', url)
```

```python
from bs4 import BeautifulSoup

html = """
<html>
        <body>
                <h1> 스크래핑이란? </h1>
                <p> 웹 페이지를 분석해서 </p>
                <p> 원하는 부분을 추출하는 기능 </p>
        </body>
</html>
"""
soup = BeautifulSoup(html, 'html.parser')

h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling

print('h1 = ' + h1.string)
print('p1 = ' + p1.string)
print('p2 = ' + p2.string)
```

```python
title = soup.find('h1')
body = soup.find('p')

print('title = ' + title.string)
print('body = ' + body.string)
```

```python
body = soup.find_all('p')

for p in body:
    print('body = ' + p.string)
```

```python
from bs4 import BeautifulSoup

html = """
<html>
        <ul>
                <li><a href='http://www.naver.com'>naver</a></li>
                <li><a href='http://www.daum.net'>naver</a></li>
        </ul>
</html>
"""

soup = BeautifulSoup(html, 'html.parser')

links = soup.find_all('a')

for a in links:
    href = a.attrs['href'] # attrs로 tag의 속성값(링크) 추출
    text = a.string
    print(text, '>', href)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req
url = 'https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'

res = req.urlopen(url)

soup = BeautifulSoup(res, 'html.parser')

title = soup.find('title').string
wf = soup.find('wf').string

print(title)
print(wf)
```

```python
import urllib.request
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/marketindex/'
res = urllib.request.urlopen(url)

soup = BeautifulSoup(res, 'html.parser')

result = soup.select_one('div.head_info > span.value')
print('usd/krw =', result.string)
```
